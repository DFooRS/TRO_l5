# Задача на вероятность
Вероятность изготовления на автоматическом станке стандартной 
детали равна 0,8. Найти вероятности возможного числа появления бракованных 
деталей среди 5 отобранных

```python
p = 0.8
q = 0.2
n = 5
C = ([np.math.factorial(n)/(np.math.factorial(k)*np.math.factorial(n - k)) for k in range (0, n + 1)])
P = ([C[i] * p**(n-i) * q**i for i in range (0, n + 1)])

labels = ["0 детелай", "1 деталей", "2 деталей", "3 деталей", "4 деталей", "5  деталей"]
fig, ax = plt.subplots()
ax.pie(P, labels=labels, rotatelabels=True)
ax.axis("equal")
```
