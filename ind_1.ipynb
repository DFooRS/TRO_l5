```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Поршневым воздушным насосом откачивают сосуд
объемом V = 100л. За один цикл (ход поршня) насос захватывает объем, равный ΔV = 0.5л. Через сколько циклов n давление в сосуде
уменьшится в $ p_0 / p_n $ раз? Процесс считать изотермическим, а
газ – идеальным.

Пусть $η_n = p_n / p_0 $ и $ q = V / (V + ΔV) $, тогда изотермический процесс можно записать в виде $ p_{n-1}V = p_n(V + ΔV)$ .
Значит, $ η_n = q^n $, а $ n = \frac{lgη_n}{lgq}$

```python
N = 950
V = 0.1
dV = 0.0005
p0 = 1.0e5

qV = V / (V + dV)
n1 = np.log10(0.01) / np.log10(qV);
print (f"n = {n1}")

x = np.empty(N) 
p = np.empty(N)
x[0] = 0
p[0] = p0
for i, item in enumerate(x):
    p0 = p0 * qV;
    x[i] = i
    p[i] = p0

plt.plot(x,p)
plt.xlabel('n')
plt.ylabel('η')
plt.show() 
```
